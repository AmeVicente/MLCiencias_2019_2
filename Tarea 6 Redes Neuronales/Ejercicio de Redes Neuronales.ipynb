```python
# Bibliotecas
import matplotlib.pyplot as plt
import pandas            as pd
import seaborn           as sns; sns.set()
%matplotlib inline

# Leemos todos los datos
datos1 = pd.read_csv("datos_1.csv")
datos2 = pd.read_csv("datos_2.csv")
datos3 = pd.read_csv("datos_3.csv")
datos4 = pd.read_csv("datos_4.csv")
datos5 = pd.read_csv("datos_5.csv")

# Graficamos nuestros puntos
for indice_temp, datos_temp in enumerate( [datos1, datos2, datos3, datos4, datos5] ):
    fig, ax1 = plt.subplots( nrows=1, ncols=1, figsize=(16,8) )
    ax1.set_title("DATOS " + str(indice_temp+1) )
    ax1.scatter( datos_temp.x, datos_temp.y, c=datos_temp.clase)
    plt.show()
```
